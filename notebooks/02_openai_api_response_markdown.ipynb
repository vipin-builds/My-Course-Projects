### Set the API KEY to Google Colab environment

```python
import os
from google.colab import userdata
from openai import OpenAI

# Securely retrieve the key from Colab Secrets
try:
    os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
    client = OpenAI()
    print("‚úÖ OpenAI Client initialized successfully.")
except Exception as e:
    print("‚ùå Error: Make sure 'OPENAI_API_KEY' is set in the Secrets tab.")
```

### Use a small model

```python
model = "gpt-5-nano"
```

### Setup messages

```python
messages = [
    {"role": "system", "content": "You are a helpful botanist expert."},
    {"role": "user", "content": "What are the 3 easiest indoor plants for a beginner to keep alive?"}
]
```

```python
# Call Open AI

response = client.chat.completions.create(
    model=model, messages=messages
)

content = response.choices[0].message.content
```

### Display the output in a formatted Markdown rather than raw text.

```python
from IPython.display import Markdown, display

display(Markdown(f"### üåø Botanist Advice\n{content}"))
```
