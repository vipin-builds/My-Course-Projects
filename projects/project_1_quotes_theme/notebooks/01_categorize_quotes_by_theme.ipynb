### Get quotes list.

### Set the API KEY to Google Colab environment

```python
import os
from google.colab import userdata
from openai import OpenAI

# Securely retrieve the key from Colab Secrets
try:
    os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')
    client = OpenAI()
    print("‚úÖ OpenAI Client initialized successfully.")
except Exception as e:
    print("‚ùå Error: Make sure 'OPENAI_API_KEY' is set in the Secrets tab.")
```

### Setup Model and Messages

```python
model = "gpt-5-nano"

content = f"""
Categorize the quotes provided into various themes.
Use broader themes like life, motivation, relationships etc..
{text_quotes}
While responding, put the data into a table format with two columns.
1) Theme name
2) Quote

And add a second table(table format here as well) which also has two columns which provides a overall theme summary of theme name and total number of quotes that belong to that theme.
1) Theme Name
2) Total
"""

messages = [
    {"role": "system", "content": "You are an assistant helping in categorizing quotes into various themes"},
    {"role": "user", "content": content}
]
```

### Make the call

```python
# Call Open AI

response = client.chat.completions.create(
    model=model, messages=messages
)

resp = response.choices[0].message.content
```

### Display the output in a formatted Markdown rather than raw text.

```python
from IPython.display import Markdown, display

display(Markdown(f"### üåø Quotes By Theme\n{resp}"))
```
